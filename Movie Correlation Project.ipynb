```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix,annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
sorted_pairs = df.corr(numeric_only=True)
                             
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

```python
correlation_matrix = df.corr(numeric_only=True)

corr_pairs = correlation_matrix.stack()

high_corr = corr_pairs[corr_pairs.abs() > 0.5].dropna()

high_corr = high_corr[high_corr != 1.0]

high_corr = high_corr.abs().sort_values(ascending=False)

print(high_corr)
```
